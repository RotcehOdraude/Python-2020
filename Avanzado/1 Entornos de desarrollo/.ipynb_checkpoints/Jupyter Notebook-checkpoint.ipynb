# Jupyter notebook

## Qu es jupyter notebook?
---
Es un cuaderno interactivo (aplicacin web) de cdigo abierto que permite crear, compartir y editar documentos en los que se puede ejecutar cdigo python, hacer anotaciones, insertar ecuaciones, visualizar resultados y documentar funcionalidades.

Esta aplicacin est diseada generalmente para tener una compatibilidad avanzada con Python, Markdown e incluye la posibilidad de exportar documentos hechos con la herramienta a otros formatos.

Generalmente esta herramienta es utilizada para el aprendizaje del lenguaje de programacin python, la limpieza y transformacin de datos cientficos, la simulacin numrica, el modelado estadstico y puede abarcar muchas otras reas. Fundamental para aquellos que desean hacer estudios de la ciencias de datos de manera ordenada, con todo el potencial de Python y con la posibilidad de dejar documentada toda la base cientfica que se ha utilizado.

Sirve a modo de puente constante entre el cdigo y los textos explicativos. De este modo, los usuarios pueden crear y compartir en tiempo real cdigo, ecuaciones, visualizaciones, etc., junto con los textos explicativos.

## Ventajas
---
+ Aunque el lenguaje de programacin fundamental en Jupyter Notebook es Python, esta aplicacin tambin es compatible con ms de 40 lenguajes, entre los que destacan R, Julia y Scala.

+ La integracin de diverso tipos de informacin nos permite dar explicaciones ms adecuada de nuestros programas o de los conceptos que estemos aprendiendo.

+ Podemos ejecutar y visualizar imgenes, videos, LaTeX y JavaScript, adems de manipular los resultados de los mismos en tiempo real.

+ El proyecto es software libre y est basado en componentes libres, por lo que est apoyado, desarrollado y mantenido por una amplia comunidad.

+ Hoy en da la herramienta est integrada en las plataformas y entornos ms importantes y extendidas como Google Cloud, Microsoft Azure, AWS, IBM Bluemix, Databricks Platform, GitHub, Rackspace, Continuum, Jetbrains

## Historia de jupyter
---
Naci como evolucin de IPython. Fue iniciado en el 2014 por el colombiano Fernando Prez PhD en Fsica de Partculas por la Universidad de Colorado y profesor en la Universidad de California en Berkeley. Sin embargo, Jupyter es mucho ms ambicioso que IPython, se pretende construir una plataforma agnstica del lenguaje que ofrezca a los cientficos un conjunto de potentes herramientas para trabajar con datos, visualizarlos y poder compartir los resultados.

El nombre Jupyter es un acrnimo de Julia, Python y R y es un homenaje a Galileo, al que se considera autor del primer paper cientfico de astronoma de la era moderna en 1610, en el cual describe sus observaciones astronmicas a travs de un telescopio de las lunas de Jpiter. Galileo demostr que la Tierra orbita el Sol de la misma manera que las lunas de Jpiter orbitan dicho planeta.



La adopcin y el xito de Jupyter ha sido muy grande, de tal modo que se est convirtiendo en un standard para el trabajo de los cientficos de datos.

## Markdown
---
Markdown es un lenguaje de marcado sencillo que sirve para agregar formato, vnculos e imgenes con facilidad al texto simple, a continuacin veremos algunos ejemplos:

* Titulos:
```Markdown
# Titulo 1
## Titulo 2
### Titulo 3
#### Titulo 4
```
* Negritas:
```Markdown
Este texto no est en negritas, **este texto si est en negritas**
```
* Italica:
```Markdown
Este texto no est en italica, *este texto si est en italica*
```
* Tachado:
```Markdown
    ~~The world is flat.~~
```
~~The world is flat.~~
* Citas largas:
```Markdown
>Don Quijote soy, y mi profesin la de andante caballera. Son mis leyes, el deshacer entuertos, prodigar el bien y evitar el mal. Huyo de la vida regalada, de la ambicin y la hipocresa, y busco para mi propia gloria la senda ms angosta y difcil. Es eso, de tonto y mentecato?.
```

* Listas:
```Markdown
    + First item
    + Second item
    + Third item
```
* Listas ordenadas:
```Markdown
    1. First item
    2. Second item
    3. Third item
```
* Codigo:
```Markdown
    `lista_de_frutas = ["Manzana","Pera","Pia","Melon","Naranja"]`
```

## Atajos de teclado
---
As any power user knows, keyboard shortcuts will save you lots of time. Jupyter stores a list of keybord shortcuts under the menu at the top: Help > Keyboard Shortcuts, or by pressing H in command mode (more on that later). Its worth checking this each time you update Jupyter, as more shortcuts are added all the time.

Another way to access keyboard shortcuts, and a handy way to learn them is to use the command palette: Cmd + Shift + P (or Ctrl + Shift + P on Linux and Windows). This dialog box helps you run any command by name  useful if you dont know the keyboard shortcut for an action or if what you want to do does not have a keyboard shortcut. The functionality is similar to Spotlight search on a Mac, and once you start using it youll wonder how you lived without it!


Some of my favorites:

+ Esc will take you into command mode where you can navigate around your notebook with arrow keys.
While in command mode:
+ A to insert a new cell above the current cell, B to insert a new cell below.
+ M to change the current cell to Markdown, Y to change it back to code
+ D + D (press the key twice) to delete the current cell
+ Enter will take you from command mode back into edit mode for the given cell.
+ Shift + Tab will show you the Docstring (documentation) for the the object you have just typed in a code cell  you can keep pressing this short cut to cycle through a few modes of documentation.
+ Ctrl + Shift + - will split the current cell into two from where your cursor is.
+ Esc + F Find and replace on your code but not the outputs.
+ Esc + O Toggle cell output.
Select Multiple Cells:
+ Shift + J or Shift + Down selects the next sell in a downwards direction. You can also select sells in an upwards direction by using Shift + K or Shift + Up.
+ Once cells are selected, you can then delete / copy / cut / paste / run them as a batch. This is helpful when you need to move parts of a notebook.
+ You can also use Shift + M to merge multiple cells.

### Multicursor support
Jupyter supports mutiple cursors, similar to Sublime Text. Simply click and drag your mouse while holding down Alt.
